```
import os
file_name='ipm_data.txt'
file_path=os.path.join("/content/gdrive/My Drive/Colab Notebooks/ex3",file_name)
```

```
import numpy as np
data=np.genfromtxt(file_path, dtype='float', delimiter='\n')
print(data.shape)
print(data.dtype)
```

```
import matplotlib.pyplot as plt
```

```
max_value=np.max(data)
max_index=np.where(data==max_value)[0]
plt.figure(figsize=(12,8))
#blue dash line
plt.plot(range(0,data.shape[0]), data, color='blue', linestyle='-', linewidth=2)
#red dot at max value
plt.plot(max_index, data[max_index],color='r', marker='o', markersize=10)
#show black grid
plt.grid(color='black')
plt.ylabel('count')
plt.xlabel('index')
plt.show()
```
