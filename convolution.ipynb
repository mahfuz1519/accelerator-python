```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = [1, 2, 3]
h = [4, 5, 6, 7]
y = np.convolve(x, h)
print(y)
```

```python
plt.stem(x)
plt.title('x')
plt.grid()
```

```python
plt.stem(h)
plt.title('h')
plt.grid()
```

```python
plt.stem(y)
plt.title('y')
plt.grid()
```
